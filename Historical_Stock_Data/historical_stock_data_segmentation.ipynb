```
def adjusted_volatility(price_list):
  price_list_after=price_list.copy()
  price_list_after.pop(0) #remove first element 
  price_list.pop() #remove last element
  return (np.mean(np.abs(pd.Series(price_list_after)-pd.Series(price_list))))
```
